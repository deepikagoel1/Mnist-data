```python
#import requisite libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn import metrics
from sklearn import datasets
import sklearn
```

```python
digits = datasets.load_digits(n_class=10) # Load the MNIST data
x_img_data = digits.data # Images
print(x_img_data)
y_img_labl = digits.target  # Corresponding labels
print(y_img_labl)
print(x_img_data.shape) #1797 imges, each saved as 64D vector
print(y_img_labl.shape)
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(x_img_data, y_img_labl, test_size = 0.3, random_state = 0)

# Splitting the Train set further into development and train set
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size = 0.2, random_state = 0)
```

```python
#Train the SVM(Gaussian kernal) 
accuracies = []
Cval = range(1,10,1)
Gval = range(1,10,1)
model = [] # list of model
modelG = [] # list of model for different degree for each Gamma
for c in range(1, 10, 1):
    Score_G = []
    for g in range(1,10,1):
        modelG.append(SVC(C=c,kernel='rbf',gamma=(g/10000)))
        modelG[g-1].fit(X_train, y_train.ravel())
        
        # evaluate the model and update the accuracies list
        score = modelG[g-1].score(X_dev,y_dev)
        Score_G.append(score)
        print("C=%d,gamma=%.5f accuracy=%.2f%%" % (c,g/10000, score * 100))
    model.append(modelG)
    accuracies.append(Score_G)


# find the value of C and Gamma that has the largest accuracy
acc = np.asarray(accuracies)
result = np.where(acc == np.amax(acc))
bestC,bestG = result[0],result[1]    
print("\n###########   best C=%d and best Gamma=%.5f achieved highest accuracy of %.2f%% on validation data" % (Cval[bestC[0]],Gval[bestG[0]]/10000,accuracies[bestC[0]][bestG[0]] * 100))


# evaluate the model on test data
C=Cval[bestC[0]]  # Get the value of best C
G=Gval[bestG[0]]   # Get the value of best Gamma
score = model[C][G].score(X_test, y_test)
print("The accuracy(%) on the test data using best C and Gamma=",score*100)


#predicting the value using Gaussian kernal for test data and finding the accuracy
Z = model[C][G].predict(X_test)
print("Accuracy =",sklearn.metrics.accuracy_score(y_test, Z)*100)


# confusion_matrix
CM=confusion_matrix(y_test, Z) 
print(CM)


# get number of support vectors for each class
print(model[C][G].n_support_)  
```
