```python
#import requisite libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn import metrics
from sklearn import datasets
import sklearn
```

```python
digits = datasets.load_digits(n_class=10) # Load the MNIST data
x_img_data = digits.data # Images
y_img_labl = digits.target  # Corresponding labels
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(x_img_data, y_img_labl, test_size = 0.3, random_state = 0)

# Splitting the Train set further into development and train set
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size = 0.2, random_state = 0)
```

```python
#Train the SVM(Polynomial Kernel)
accuracies = []
Cval = range(1,10,1)
Dval = range(1,10,1)
model = [] # list of model
modelD = [] # list of model for different degree for each Degree
for c in range(1, 10, 1):
    Score_D = []
    for d in range(1,10,1):
        modelD.append(SVC(C=c,kernel='poly',degree=d))
        modelD[d-1].fit(X_train, y_train.ravel())
        
        # evaluate the model and update the accuracies list
        score = modelD[d-1].score(X_dev,y_dev)
        Score_D.append(score)
        print("C=%d,D=%d accuracy=%.2f%%" % (c,d, score * 100))
    model.append(modelD)
    accuracies.append(Score_D)


# find the value of C and Degree that has the largest accuracy 
acc = np.asarray(accuracies)
result = np.where(acc == np.amax(acc))
bestC,bestD = result[0],result[1]   
print("\n###########   best C=%d and best D=%d achieved highest accuracy of %.2f%% on validation data" % (Cval[bestC[0]],Dval[bestD[0]],accuracies[bestC[0]][bestD[0]] * 100))


# evaluate the model on test data
C=Cval[bestC[0]]  # Get the value of best C
D=Dval[bestD[0]]   # Get the value of best Degree
score = model[C][D].score(X_test, y_test)
print("The accuracy(%) on the test data using best C =",score*100)


#predicting the value using Polynomial kernal for test data and finding the accuracy
Z = model[C][D].predict(X_test)
print("Accuracy =",sklearn.metrics.accuracy_score(y_test, Z)*100)


# confusion_matrix
CM=confusion_matrix(y_test, Z) 
print(CM)


# get number of support vectors for each class
print(model[C][D].n_support_)  
```
